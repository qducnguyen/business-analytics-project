```python
with open("data_json.json") as fIn:
    data_json_fixed = json.load(fIn)
```

```python
data_json_fixed
```

```python
data_tourism["province_eng"] = data_tourism["province_eng"].map(data_json_fixed)
```

```python
data_tourism
```

```python
# Get only english data
df_tourism = data_tourism[['date', 'province_eng', 'region_eng', 'variable', 'value']] \
    .rename(columns={'province_eng': 'province', 'region_eng': 'region'})
df_tourism
```

```python
list(data_tourism["variable"].unique())
```

```python
index_now = df_tourism.index[(df_tourism["variable"] == "ratio_tourist_stay")].tolist()
df_tourism.loc[index_now, "value"] = df_tourism.loc[index_now, "value"]  / 100
```

```python
# Flatten variable column
df_tourism
```

```python

data_json = {}
for _, row in df_tourism.iterrows():
    date = row["date"]
    province = row["province"]
    region = row["region"]
    variable_name = row["variable"]
    variable_value = row["value"]
    key = (date, province, region)
    if key not in data_json:
        data_json[key] = {}
    data_json[key][variable_name] = variable_value
```

```python
data_json
```

```python
data_csv = []
for key, value in data_json.items():
    data_sample = {"date": key[0], "province": key[1], "region": key[2]}
    for k, v in value.items():
        data_sample[k]  = v
    data_csv.append(data_sample)
```

```python
len(data_csv)
```

```python
pd_new = pd.DataFrame.from_dict(data_csv)
```

```python
pd_new
```

```python
pd_new.isnull().sum()
```
